# Import Necessary Module

```python
import pandas as pd
import matplotlib.pyplot as plt
import io
import numpy as np
```

# Upload Csv File

```python
from google.colab import files
from datetime import datetime


uploaded = files.upload()
filename = next(iter(uploaded))
df = pd.read_csv(filename, parse_dates=['Datetime'])
```

# Plot Data 

```python
print("Data Shape >>", df.shape)
print(df.columns)
df.head()
#df['Datetime'][0]
```

```python
# Extract data for plotting 
y_label = 'Exposure Time(ms)'
z_label = 'Total Time(ms)' 
```

```python
# Starting the Plot
y = df[y_label]
z = df[z_label]
# x = np.arange(len(y))
x = df["Datetime"]        # For x Axis data

# Create a figure and axis
fig, ax = plt.subplots()

# Plot the data
'''
easy_color =  ['r', 'g', 'b', 'c', 'm', 'k']
Color = #FFA500 (สีส้มเหลือง)
'''

ax.plot(x, y, color = 'b', linestyle = 'dashed',
       label= y_label)                      #  marker = '.',  label = "xxxx"
ax.plot(x, z, color = '#FFA500', linestyle = 'solid',
       label= z_label)                      #  marker = '.',  label = "xxxx"

# Set labels and title
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_title("CSV Data Plot")
ax.legend()             # Legend Location

# Show the plot
#plt.show()
plt.savefig('figure.png')
files.download('figure.png')
```
